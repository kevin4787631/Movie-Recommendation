```python
movie_id = movie_dataframe.id.tolist()
movie_homepage = movie_dataframe.homepage.tolist()
movie_title = movie_dataframe.title_x.tolist()
movie_genres = movie_dataframe.genres
movie_overview = movie_dataframe.overview.tolist()
movie_keywords = movie_dataframe.keywords
movie_cast = movie_dataframe.cast
movie_crew = movie_dataframe.crew
```

```python
new_dataframe = pd.DataFrame({
    'id': movie_id,
    'title': movie_title,
    'homepage': movie_homepage,
    'genres' : movie_genres,
    'overview': movie_overview,
    'runtime' : movie_dataframe.runtime,
    'vote_average' : movie_dataframe.vote_average,
    'vote_count': movie_dataframe.vote_count,
    'release_date' : movie_dataframe.release_date,
    'keywords': movie_keywords,
    'cast': movie_cast,
    'crew': movie_crew
})
```

```python
new_dataframe
```

```python
import ast

def dict_conversion(dict):
    list=[]
    for i in ast.literal_eval(dict):
        list.append(i['name'])
    return list
```

```python
new_dataframe['genres'] = new_dataframe['genres'].apply(dict_conversion)
new_dataframe['keywords'] = new_dataframe['keywords'].apply(dict_conversion)
```

```python
def dict_conversion_3(dict):
    list=[]
    count = 0
    for i in ast.literal_eval(dict):
        if count != 3:
            list.append(i['name'])
            count += 1
        else:
            break
    return list
```

```python
new_dataframe['cast'] = new_dataframe['cast'].apply(dict_conversion_3)
```

```python
def get_director(dict):
    list = []
    for i in ast.literal_eval(dict):
        if i['job'] == 'Director':
            list.append(i['name'])
            break
    return list
```

```python
new_dataframe['crew'] = new_dataframe['crew'].apply(get_director)
```

```python
new_dataframe['overview'] = new_dataframe['overview'].apply(lambda x:x.split())
```

```python
new_dataframe['genres'] = new_dataframe['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
new_dataframe['cast'] = new_dataframe['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
new_dataframe['crew'] = new_dataframe['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
new_dataframe['keywords'] = new_dataframe['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stemming(string):
  list = []
  for i in string.split():
    list.append(ps.stem(i))
  return " ".join(list)
```

```python
new_dataframe['tags'] = new_dataframe['tags'].apply(stemming)
```

```python
new_dataframe.head(3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vector_count = CountVectorizer(max_features = 5000, stop_words='english')
```

```python
vectors = vector_count.fit_transform(new_dataframe['tags']).toarray()
```

```python
print(vectors)
```

```python
vectors.shape
```

```python
vector_count.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
print(similarity)
```

```python
def rec(movie):
  index = new_dataframe[new_dataframe['title'] == movie].index[0]
  movie_genre = new_dataframe.iloc[index].genres
  dist = similarity[index]
  movies_list = sorted(list(enumerate(dist)), reverse=True, key = lambda x:x[1])[1:11]
  total_percentage = 0
  for i in movies_list:
    count = 0
    genre = new_dataframe.iloc[i[0]].genres
    for j in genre:
      if j in movie_genre:
        count += 1
    percentage = int(count/len(movie_genre)*100)

    print(new_dataframe.iloc[i[0]].title, '\nCosine Similarity: {}'.format(i[1]), '\n')
    total_percentage += percentage

  average_precision = total_percentage/10
  print("The genre precision at 10 recommendations for movie '{}': {}%".format(movie, average_precision))

  return
```

```python
rec('The Avengers')
```

#Evaluation (MAP@K)

```python
map = 0
for movie in new_dataframe['title']:
  num_of_recs = 10
  index = new_dataframe[new_dataframe['title'] == movie].index[0]
  movie_genre = new_dataframe.iloc[index].genres
  dist = similarity[index]
  movies_list = sorted(list(enumerate(dist)), reverse=True, key = lambda x:x[1])[1:num_of_recs+1]
  total_percentage = 0
  for i in movies_list:
    count = 0
    genre = new_dataframe.iloc[i[0]].genres
    for j in genre:
      if j in movie_genre:
        count += 1

    if len(movie_genre) != 0:
      percentage = int(count/len(movie_genre)*100)
    else:
      percentage = 0
    total_percentage += percentage

  average_precision = total_percentage/num_of_recs
  print("The genre precision at {} recommendations for movie '{}': {}%".format(num_of_recs, movie, average_precision))

  map += average_precision

map = map/len(new_dataframe['title'])
print("The Mean Average Precision of genre at {} recommendations is: {}".format(num_of_recs, map))
```
